```python
def dateDiff(date1, date2):
  ##
  # Python's program to calculate time difference between two datetime objects.

  datetimeFormat = '%Y-%m-%dT%H:%M:%S'
  #example format time
  #date1 = '2021-02-02T22:10:00+02:00'
  #date2 = '2021-02-02T22:00:00+02:00'

  size = len(date1)
  size2 = len(date2)

  # Slice string to remove last 6 characters from string
  date1 = date1[:size - 6]
  date2 = date2[:size2 - 6]

  diff = datetime.datetime.strptime(date1, datetimeFormat)\
    - datetime.datetime.strptime(date2, datetimeFormat)
  #print("Seconds:", diff.seconds)
  return diff.seconds
```

```python
#Function of the distance between two coordinatesS
def coorDist(latO,lonO,latP,lonP):
  origin = (latO,lonO)
  dist = (latP,lonP)
  return geodesic(origin, dist).meters
```

```python
#in this forloop we are searching the time between station.
#ArrivalTime - this is time between station of the Misrad Tahbura model
#StationTime - this is time vetween station which we are finding, we are taking RecordedTime - DepurtureTime + tikun to the station.
for i in range(df2.shape[0]):
  data1 = df2['RecordedAtTime'][i]
  data2 = df2['OriginAimedDepartureTime'][i]
  data3 = df2['ExpectedArrivalTime'][i]
  k = df2['StopPointRef'][i]
  arr = dfBus.loc[[k][0]]
  x = coorDist(arr[1],arr[2],df2['Latitude'][i],df2['Longitude'][i])

  df2['StationTime'][i] = dateDiff(data1, data2) + (x/(avrgVel/3.6))
  df2['ArrivalTime'][i] = dateDiff(data3, data2)
```

#Converting time to hours, minutes seconds

```python
#Featerus for RecordetAtTime column
df2['A_Hours'] = 0
df2['A_Minutes'] = 0
df2['A_Seconds'] = 0
#Featerus for OriginAimedDepartureTime 
df2['D_Hours'] = 0
df2['D_Minutes'] = 0
#feature for weeday name
df2['WeekDay'] = 'NA'
df2.head(12)
```

```python
#Spliting data for departure time column
for i in range(df2.shape[0]):
  hours = df2['OriginAimedDepartureTime'][i]
  minutes = df2['OriginAimedDepartureTime'][i]
  sec = df2['OriginAimedDepartureTime'][i]
  size = len(hours)
  hours = hours[:size - 12]
  hours = hours[11:]
  minutes = minutes[:size - 9]
  minutes = minutes[14:]
  sec = sec[:size - 6]
  sec = sec[17:]
  df2['D_Hours'][i] = int(hours)
  df2['D_Minutes'][i] = int(minutes)
df2.drop('OriginAimedDepartureTime', axis='columns', inplace=True)
df2
```

```python
#Spliting data for RecordedAtTime column and converting the date weekday
for i in range(df2.shape[0]):
  hours = df2['RecordedAtTime'][i]
  minutes = df2['RecordedAtTime'][i]
  sec = df2['RecordedAtTime'][i]
  year = df2['RecordedAtTime'][i]
  month = df2['RecordedAtTime'][i]
  day = df2['RecordedAtTime'][i]
#Spliting for hours, minutes, seconds
  size = len(hours)
  hours = hours[:size - 12]
  hours = hours[11:]
  minutes = minutes[:size - 9]
  minutes = minutes[14:]
  sec = sec[:size - 6]
  sec = sec[17:]
#Spliting the date and find the weekday
  year = year[:size - 21]
  year = year[0:]
  month = month[:size - 18]
  month = month[5:]
  day = day[:size - 15]
  day = day[8:]
  now = datetime.datetime(int(year),int(month),int(day))
  df2['WeekDay'][i] = now.strftime("%A") 
  df2['A_Hours'][i] = int(hours)
  df2['A_Minutes'][i] = int(minutes)
  df2['A_Seconds'][i] = int(minutes)
df2.drop('RecordedAtTime', axis='columns', inplace=True)
df2
```

```python
#fixing the columns with column stationtime
for i in range(df2.shape[0]):
  my_time_m = int(df2['StationTime'][i]/60)
  my_time_s = int(df2['StationTime'][i]%60)
  df2['A_Hours'][i] = int(df2['A_Hours'][i])
  df2['A_Minutes'][i] = int(df2['A_Minutes'][i]) + my_time_m
  df2['A_Seconds'][i] = int(df2['A_Seconds'][i]) + my_time_s
```

```python
#fixing the seconds column if the data more then 60
for i in range(df2.shape[0]):
  if df2['A_Seconds'][i] >=60:
    #print(i)
    df2['A_Minutes'][i] = df2['A_Minutes'][i] +1
    df2['A_Seconds'][i] = df2['A_Seconds'][i] - 60
```

```python
#fixing the minutes column if the data more then 60
for i in range(df2.shape[0]):
  if df2['A_Minutes'][i] >=60:
    #print(i)
    df2['A_Hours'][i] = df2['A_Hours'][i] +1
    df2['A_Minutes'][i] = df2['A_Minutes'][i] - 60
```
